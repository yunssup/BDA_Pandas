### **Pandas 기본 개념 정리**

Pandas는 **데이터 분석**과 **조작**을 위한 Python 라이브러리로, **데이터프레임(DataFrame)**을 활용해 데이터를 효율적으로 처리할 수 있습니다. 주로 **구조화된 데이터**(표 형식)를 다루며, CSV, Excel, SQL 등 다양한 데이터 소스에서 데이터를 가져와 분석할 수 있습니다.

---

### 1. **Pandas의 주요 데이터 구조**

#### 1.1 **Series**
- **정의**: Pandas에서 제공하는 **1차원 데이터** 구조로, **인덱스**와 **값**으로 구성된 데이터 타입입니다.
- **특징**:
  - NumPy 배열과 유사하지만, **인덱스**라는 레이블을 사용하여 데이터에 쉽게 접근할 수 있습니다.
  - **정수, 문자열, 부동 소수점 숫자** 등 다양한 데이터 타입을 저장할 수 있습니다.

#### 1.2 **DataFrame**
- **정의**: **2차원 데이터 구조**로, 행(row)과 열(column)로 구성된 **표 형식의 데이터**입니다.
- **특징**:
  - 각 열은 **서로 다른 데이터 타입**을 가질 수 있으며, 행과 열 모두 **인덱스**를 사용해 접근할 수 있습니다.
  - 엑셀이나 SQL 테이블과 유사한 형식으로 데이터를 저장하고 분석하는 데 최적화되어 있습니다.

#### 1.3 **Index**
- **정의**: **Series**와 **DataFrame**에서 데이터의 **행**과 **열**을 참조하는 **레이블**입니다.
- **특징**:
  - 기본적으로 Pandas는 0부터 시작하는 **정수 인덱스**를 자동으로 할당하지만, 사용자가 직접 지정할 수도 있습니다.
  - 인덱스를 사용해 데이터의 특정 행이나 열에 쉽게 접근할 수 있습니다.

---

### 2. **Pandas의 주요 기능**

#### 2.1 **데이터 읽기/쓰기**
- Pandas는 다양한 **데이터 소스**로부터 데이터를 **불러오고 저장**할 수 있는 기능을 제공합니다. CSV, Excel, SQL 등과 같은 형식에서 데이터를 쉽게 불러올 수 있으며, 변환된 데이터를 다시 저장할 수 있습니다.

#### 2.2 **데이터 선택 및 필터링**
- Pandas는 데이터를 선택하고 필터링하는 다양한 방법을 제공합니다.
  - **인덱싱**과 **슬라이싱**을 통해 데이터의 특정 부분을 선택할 수 있습니다.
  - **조건을 사용한 필터링**을 통해 원하는 조건을 만족하는 데이터를 쉽게 추출할 수 있습니다.

#### 2.3 **데이터 정렬**
- Pandas는 데이터를 **특정 열**이나 **인덱스**를 기준으로 정렬할 수 있습니다. 오름차순이나 내림차순으로 정렬하는 기능을 제공하여 데이터 분석을 더욱 쉽게 할 수 있습니다.

#### 2.4 **결측값 처리**
- 데이터셋에는 종종 **결측값**(Missing Values)이 존재할 수 있는데, Pandas는 이를 처리하는 다양한 방법을 제공합니다.
  - 결측값을 **제거**하거나, **대체 값**을 넣는 방식으로 처리할 수 있습니다.

#### 2.5 **데이터 변형**
- Pandas는 데이터를 변형하거나 조작하는 다양한 기능을 제공합니다.
  - **열 추가/삭제**: 새로운 열을 추가하거나 불필요한 열을 제거할 수 있습니다.
  - **데이터 형식 변환**: 데이터 타입을 쉽게 변환할 수 있으며, 이를 통해 메모리 최적화를 할 수 있습니다.

#### 2.6 **그룹화와 집계**
- Pandas는 데이터를 특정 기준으로 **그룹화**하고, 그룹별로 **집계 연산**(sum, mean 등)을 수행하는 기능을 제공합니다. 이를 통해 데이터의 요약 통계를 계산하거나 특정 패턴을 분석할 수 있습니다.

#### 2.7 **병합과 결합**
- Pandas는 두 개 이상의 **데이터프레임**을 **병합**(merge)하거나 **결합**(concatenate)하는 기능을 제공합니다. SQL의 JOIN과 유사한 방식으로 데이터를 병합할 수 있으며, 행 또는 열 방향으로 데이터를 결합할 수 있습니다.

#### 2.8 **데이터 프레임의 통계 분석**
- Pandas는 데이터의 **기본 통계값**(평균, 표준편차, 최소값/최대값 등)을 계산하는 기능을 제공합니다. 이를 통해 데이터의 분포를 빠르게 파악할 수 있습니다.

### 3. **Pandas의 장점**
- **사용 용이성**: Pandas는 직관적인 API를 제공하여, 복잡한 데이터 조작도 쉽게 할 수 있습니다.
- **데이터 호환성**: CSV, Excel, SQL 등 다양한 포맷의 데이터를 불러오고 저장할 수 있습니다.
- **다양한 기능**: 데이터 필터링, 변환, 결합, 집계 등 데이터 분석에 필요한 모든 기능을 제공.
- **빠른 데이터 처리**: Pandas는 내부적으로 **NumPy**를 사용하여 **고속**의 데이터 처리를 지원합니다.

1. Pandas와 SQL 연동 개념
SQLAlchemy: Python에서 다양한 데이터베이스(MySQL, PostgreSQL, SQLite 등)에 연결할 수 있는 라이브러리입니다. Pandas는 SQLAlchemy를 통해 데이터베이스에 연결하고, SQL 쿼리를 실행하여 데이터를 가져올 수 있습니다.
pandas.read_sql_query(): SQL 쿼리를 실행하고 그 결과를 Pandas의 DataFrame으로 불러옵니다.
pandas.to_sql(): Pandas DataFrame을 SQL 테이블로 저장합니다.

```python
import pandas as pd
import seaborn as sns #데이터 시각화 도구
from sqlalchemy import create_engine

titanic = sns.load_dataset('titanic')
#SQLite 데이터베이스 연결 설정 (SQLite 사용)
engine = create_engine('sqlite:///titanic.db', echo=False)
# 판다스 데이터프레임을 SQL 테이블로 저장 (method='multi'추가)
titanic.to_sql('titanic', con=engine, if_exists='replace', index=False, method='multi')
# 쿼리 사용해 데이터 읽기
df_from_sql=pd.read_sql_query('SELECT * FROM titanic', engine)
print(df_from_sql.head())
```
